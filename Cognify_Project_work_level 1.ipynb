# Fetching the Top 3 cuisines 

```python
#spliting the cousines and counting the frequency
cuisine_counts = df['Cuisines'].str.split(', ').explode().value_counts()
```

```python
#top 3 most common cuisines 
top_cuisines = cuisine_counts.head(3)
```

# % of restuarant that serve the most common cuisines

```python
#calculating the % of restaurant serving each of top cuisines
total_restaurants = len(df)
top_cuisines_percentage = (top_cuisines / total_restaurants) * 100
```

```python
top_cuisines, top_cuisines_percentage
```

## ANSWERS:

#### North Indian with 3,960 restaurants serving this cuisine, which constitutes approximately 41.46% of all restaurants in the dataset.    
#### Chinese with 2,735 restaurants serving this cuisine, making up about 28.64% of the restaurants.    
#### Fast Food with 1,986 restaurants, accounting for 20.79% of the total number of restaurants.

```python
# using a bar chart to visualizing the above task
```

```python
cuisines = top_cuisines.index
percentages = top_cuisines_percentage.values

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(cuisines, percentages, color=['green', 'orange', 'yellow'])
plt.title('Top 3 Most Common Cuisines and Their Percentage')
plt.xlabel('Cuisine')
plt.ylabel('Percentage of Restaurants (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.show()
```

# <span style="color:GREEN">CITY ANALYSIS:::</span> 

#### 1. Identify the city with the highest number
#### of restaurants in the dataset.
#### 2. Calculate the average rating for
#### restaurants in each city.
#### 3. Determine the city with the highest
#### average rating

```python
#INDENTIFYING THE CITY WITH HIGHEST NUMBNER OF RESTAURANT

city_counts = df['City'].value_counts().head(1)
```

```python
# CALCULATE THE AVERAGE RATING FOR RESTAURANT IN EACH CITY

average_rating_per_city = df.groupby('City')['Aggregate rating'].mean()
```

```python
# CITY WITH HIGHEST AVERAGE RATING

city_highest_average_rating = average_rating_per_city.idxmax()
highest_average_rating = average_rating_per_city.max()

city_counts, city_highest_average_rating, highest_average_rating
```

## ANSWERS:

#### The city with the highest number of restaurants in the dataset is New Delhi, with a total of 5,473 restaurants.

#### The average rating for restaurants has been calculated for each city within the dataset.

#### The city with the highest average rating for its restaurants is Inner City, where the restaurants have an average rating of 4.9. 

```python
#visualize the top cities by average rating for clarity
top_average_ratings = average_rating_per_city.sort_values(ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 8))
top_average_ratings.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities by Average Restaurant Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Highlight the city with the highest average rating
plt.scatter(x=[0], y=[highest_average_rating], color='red', zorder=5, label='Highest Average Rating (Inner City)')
plt.legend()

# Show the plot
plt.show()
```

#### The bar chart above visualizes the top 10 cities by average restaurant rating within the dataset, with a special highlight on Inner

#### City, which has the highest average rating of 4.9. This visualization effectively showcases the cities with the best-rated 

#### restaurants, according to the dataset, emphasizing the outstanding quality of dining options available in Inner City.

# <span style="color:GREEN"> ONLINE DELIEVERY::</span> 
1. Determine the percentage of restaurants
that offer online delivery.

2. Compare the average ratings of restaurants
with and without online delivery.

```python

# DetermineD the percentage of restaurants that offer online delivery
online_delivery_counts = df['Has Online delivery'].value_counts(normalize=True) * 100
```

```python
# Compare the average ratings of restaurants with and without online delivery
average_rating_with_online_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_online_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()

online_delivery_counts, average_rating_with_online_delivery, average_rating_without_online_delivery
```

#### 74.34% of restaurants in the dataset do not offer online delivery, while 25.66% of restaurants do offer this service.
#### When comparing the average ratings of restaurants based on their online delivery status:

##### * Restaurants with online delivery have an average rating of 3.25.
##### * Restaurants without online delivery have a lower average rating of 2.47.

###### This analysis indicates that a significant majority of restaurants do not offer online delivery. Additionally, restaurants that do offer online delivery tend to have higher average ratings compared to those that do not provide this service, suggesting a possible correlation between offering online delivery and receiving higher customer ratings
